 #### Concept 2 — dictionary par loop (sabse common pattern)

```python
new_dic = {
           "laptop":7 , 
           "mouse" : 9,
           "keyboard" : 10 ,
           "board" : 12
}

print(new_dic)
```

```python
for key,value in new_dic.items() :
    print(f"I bought {key} = {value}")
```

```python

for key in new_dic.keys():
    print(f"your key is {key}")
```

```python
for value in new_dic.values():
    print(f"your key value is {value}")
```

## Concept 2 — set operations (maths jaise)

#### union (|) , intersection (&) , diffrence (-)

```python
a = {9,7,6,3,7,6,4,3,0,4}
b = {9,8,5,0,3,2,1,0,4,6}
```

```python
a
```

```python
b
```

```python
a_b_union = a | b
print(a_b_union)
```

```python
a_b_intersection = a & b
print(a_b_intersection)
```

```python
a_diff_b = a - b
print(a_diff_b)
```

```python
b_diff_a = b - a
print(b_diff_a)
```
