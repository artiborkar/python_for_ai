```python
##Do boolean variables banao aur unpar and, or, not ka result print karo

var1=True
var2=False
print(f"bollean value is {var1 and var2}")            #ek value false hai to print false hoga
print(f"bollean value is {var1 or var2}")              #ek bhi value true hai to print true hoga
print(f"bollean value is {not var2}")                   #not ye reverse karta hai value ko 

```
