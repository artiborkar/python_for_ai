3. Marks lo aur upar wala grade A/B/C/D print karo

```python
marks =int(input("Enter Your Marks : "))

if marks >= 80 :
    print("Grade A")
elif marks >= 70 and marks < 80 :
    print("Grade B")
elif marks >= 60 and marks < 70 :
    print("Grade C")
elif marks >= 50 and marks < 60 :
    print("Grade D")
```
