 Concept 4 — zip (do lists saath-saath)

```python
items=['laptop','mouse','keyboard','monitor']
items_count=[4,3,2,1]

for item_count , item in zip(items_count , items):
    print(f"{item_count} {item}")
```

```python
items=['laptop','mouse','keyboard','monitor']
items_count=[4,3,2,1]
items_brand = ['apple','dell','hp','asue']

for item_count , item ,item_brand in zip(items_count , items,items_brand):
    print(f"{item_count} {item} {item_brand}")
```
