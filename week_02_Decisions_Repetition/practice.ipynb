Ek program: number lo, print karo "Positive", "Negative", ya "Zero".

```python
num = int(input("Enter the number:"))

if num > 0 :
    print(f" The number is positive : {num}")
elif num < 0 :
    print(f"The number is negative:{num}")
else :
    print(f"The number is zero : {num }")
```

Marks lo aur upar wala grade A/B/C/D print karo

```python
marks = int (input ("Enter your Marks : "))

if marks >= 90 :
    print(f"Your Grade is A :{marks }")

elif marks >= 70 and marks <= 90 :
    print(f"Your Grade is B :{marks }")

elif marks >= 50 and marks <= 70 :
    print(f"Your Grade is C :{marks }")

else :
    print(f"Your Grade is D :{marks }")
```

4.Teen numbers lo aur sabse bada print karo (if/elif/else se).

```python
num1 = int(input("Enter the Number one : "))
num2 = int(input("Enter the Number Two : "))
num3 = int(input("Enter the Number Three : "))

if num1 > num2 and num1 > num3 :
    print(f"Largest Number is num1 : {num1}")
elif num2 > num3 :
    print(f"Largest Number is num 2 : {num2}")
else :
    print(f"Largest Number is  num3 : {num3}")
```

Ek list ["red","green","blue"] ko enumerate se number ke saath print karo.

```python
colors = ["red","green","blue"]

for index,color in enumerate(colors,start=2):
    print(f"{index}.{color}")
```

User se numbers maangte raho jab tak woh "stop" na likhe; phir total sum print karo

```python
total=0
while True :
    user = input("Enter the number : ")
    if user == "stop" :
        break
    total=total+int(user)
print(f"total={total}")
```
